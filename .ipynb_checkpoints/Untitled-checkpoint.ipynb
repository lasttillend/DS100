```python
def average(n):
    """Return the average of all arguments ever passed to the average function.
    
    >>> average(1)
    1.0
    >>> average(3)
    2.0
    >>> average(8)
    4.0
    >>> average(0)
    3.0
    """
    # BEGIN SOLUTION
    average.total += n
    average.k += 1
    return average.total / average.k
average.total, average.k = 0, 0
```

```python
def f(t):
    return 3 * np.sin(2 * np.pi * t)

t = np.linspace(0, np.pi, 30)
plt.plot(t, f(t), 'r+')
plt.axis([0, np.pi, -10, 10])
plt.xlabel('t')
plt.ylabel('f(t)')
plt.xticks([0, np.pi/2, np.pi], [r'$0$', r'$\pi/2$', r'$\pi$'])
plt.title(r'$f(t) = 3\sin(2\pi t)$')
plt.grid(False)
```
